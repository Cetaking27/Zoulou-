## Задание 1:
Построить матрицу c количеством строк 10 и столбцов 2, переменные 
заполняются случайными числами с равномерным распределением. 
Полученные сгенерированные случайные числа представить на графике в 
виде точек. 

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
matrix_a = np.random.uniform(size=(10,2))
matrix_a
```

```python
plt.figure(figsize=(5,5))
plt.plot(matrix_a[:,0], matrix_a[:,1],"y^")
plt.show()
```

## Задание 2: 
Сгенерировать 1000 случайных чисел с любым известным распределением 
и построить их гистограмму, математическое ожидание, дисперсию.


```python
mat_a = np.random.exponential(size= 1000)
x = np.mean(mat_a)
z = np.var(mat_a)
print(mat_a)
print(x)
print(z)

```

```python
from matplotlib.pyplot import axis


plt.figure(figsize=(5,5))
plt.hist(mat_a, density=True, histtype='stepfilled')
plt.hist(x, density=True, histtype='stepfilled' )
plt.hist(z, density=True, histtype='stepfilled')
plt.show()
```

## Задание 3 :
Сгенерировать случайную точку, равномерно распределенную в квадрате 
со стороной а. 

```python
from matplotlib.pyplot import xlim


a = (1,2)
number = np.random.uniform(a)
print(number)

plt.hist(a, bins=(10))
plt.xlim(number)
plt.show()
```
