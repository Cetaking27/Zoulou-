```python
import numpy as np 
import matplotlib.pyplot as plt 
import random 
import math
```

```python
#PCA метод главный компонет
#шаг 1 подготовка данный 
x = np.arange(1,11)
y = 2*x + np.random.randn(10)*2
X = np.vstack((x,y))
X
```

```python
Xcentered = (X[0] - x.mean(), X[1] - y.mean())
m = (x.mean(), y.mean())
print (Xcentered) 
print("Mean vector:", m)
```

```python
#шаг 2 ковариационнная матрица 
covmat = np.cov(Xcentered) #значение коварационая мартрица
print(covmat, "\n")
print("var of X: ", np.cov(Xcentered)[0,0])
print("var of y: ", np.cov(Xcentered)[1,1])
print("var X and y: ", np.cov(Xcentered)[0,1])
```

```python
_, vecs = np.linalg.eig(covmat)
v = -vecs[:,1]
Xnew = np.dot(v,Xcentered)
print(Xnew)
```

```python
n = 9
Xcentered = np.dot(Xnew[n],v) + m
print("restored: ", Xcentered)
print("original: ", X[:,n])
```
